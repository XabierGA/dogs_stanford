```python
pixels = 100
rgb = 3
dim = (pixels, pixels)
```

```python
def parse_dir(image_address):
    dir_image = os.listdir(image_address)
    m = 0
    for race in dir_image:
        race_dir = os.listdir(image_address + '/' + race)
        for img in race_dir:
            m += 1
            
    tensor = np.zeros((m , pixels, pixels , rgb), dtype = np.uint8)
    
    return tensor , m
```

```python
def parse_img(img_route , dim):
    
    img = cv2.imread(img_route)
    dim = tuple(dim)
    img = cv2.resize(img, dim , interpolation = cv2.INTER_AREA)
    
    return img
    
```

```python
def fill_tensor(image_address , dim):

    dir_image = os.listdir(image_address)
    
    n = len(dir_image)
    
    tensor , m = parse_dir(image_address)
    print(m)
    
    Y = np.zeros((m , n) , dtype = np.uint8)
        
    img_counter = 0
    
    for j , race in enumerate(dir_image): 
        
        race_dir = os.listdir(image_address + '/' + race)
        
        factor = img_counter
        
        for i,img in enumerate(race_dir):
            
            tensor[i+factor] = parse_img(image_address + '/' + race + '/' + img , dim)
            #print(tensor[i+n*j])
            Y[img_counter][j] = 1
            
            img_counter += 1
            
            
    return tensor , Y
```
